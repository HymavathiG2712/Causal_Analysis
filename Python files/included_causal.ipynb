```python
import json
```

```python
import jsonlines

file1 = 'causenet-precision.jsonl'
rel = []
sentences = []

with open(file1, 'r') as f:
    for lineno, line in enumerate(f, start=1):
        try:
            data = json.loads(line)
            cause = data['causal_relation']['cause']['concept']
            effect = data['causal_relation']['effect']['concept']
            try:
                sen = data['sources'][0]['payload']['sentence']
            except:
                sen = ''
            rel.append((cause, effect))
            sentences.append(sen)
        except json.JSONDecodeError:
            print(f"Error decoding JSON on line {lineno}. Skipping...")
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(zip(sentences, rel)), columns =['sentences','relation'])

df = df[df['sentences']!='']
df = df.reset_index()

df
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

causal_df = pd.read_csv('total_causal.csv')
non_causal_df = pd.read_csv('non_causal_df.csv')

causal_train, causal_test = train_test_split(causal_df, test_size=0.3, random_state=42)
non_causal_train, non_causal_test = train_test_split(non_causal_df, test_size=0.3, random_state=42)

train_dataset = pd.concat([causal_train, non_causal_train], ignore_index=True)
test_dataset = pd.concat([causal_test, non_causal_test], ignore_index=True)

train_dataset = train_dataset.sample(frac=1).reset_index(drop=True)
test_dataset = test_dataset.sample(frac=1).reset_index(drop=True)
```
